```python
X = addata[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male']]
Y = addata['Clicked on Ad']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=101)
```

```python
window = Tk()
window.title('Efficient Advertisement')
window.config(padx = 50, pady = 50)
lr = LogisticRegression()
lr.fit(X_train, Y_train)

def yon():
    pred_y = lr.predict(X_test)
    z = lr.predict([[float(input1.get()),int(input2.get()),float(input3.get()),float(input4.get()),int(input5.get())]])
    if z==0:
        label5.config(text ="User won't click on ad")
    else:
        label5.config(text ="User will click on ad")


label = Label(text = 'Daily Time Spent on Site')
label.grid(row=0,column=0)

input1 = Entry()
input1.grid(row=0,column=1)

label1 = Label(text = 'Age')
label1.grid(row=1,column=0)

input2 = Entry()
input2.grid(row=1,column=1)

label2 = Label(text = 'Area Income')
label2.grid(row=2,column=0)

input3= Entry()
input3.grid(row=2,column=1)

label3 = Label(text= 'Daily Internet Usage')
label3.grid(row=3,column=0)

input4 = Entry()
input4.grid(row=3,column=1)

label4 = Label(text = 'Male')
label4.grid(row=4,column=0)

input5 = Entry()
input5.grid(row=4,column=1)

but = Button(text='Check',command=yon)
but.grid(row=5,column=1)

label5 = Label(text='',font=('Newsreader',30),fg='green')
label5.grid(row=6,column=1)

window.mainloop()
```
